```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Cars93.csv')
df
```

```python
df.shape 
```

```python
df.head(10)
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.boxplot(df['Width'])
plt.subplot(1,2,2)
sns.boxplot(df['Weight'])
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df['Width'],color='c')
plt.subplot(1,2,2)
sns.distplot(df['Weight'],color='r')
plt.show()
```
