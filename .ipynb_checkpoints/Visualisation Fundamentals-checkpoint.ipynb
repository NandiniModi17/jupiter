# Why visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame([[10,20],[11,25],[12,21],[9,18],[8,18]],index=[1,2,3,4,5],columns=['PP','DSP'])
r=df.index
pp=df['PP']
dsp=df['DSP']
plt.plot(r,pp)
plt.plot(r,dsp)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,40,30,80,40]
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,40,30,80,40]
plt.plot(x,y)
plt.title("First Chart",color='k',fontsize=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,40,30,80,40]
plt.plot(x,y)
plt.title("One Subject Marks",color='black',fontsize=26)
plt.xlabel("Roll Number")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,40,30,80,40]
plt.plot(x,y)
plt.title("One Subject Marks",color='r',fontsize=25)
plt.xlabel("Roll Number")
plt.ylabel("Marks")
plt.ylim([0,100])
plt.xlim([1,10])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[10,40,30,80,40]
plt.plot(x,y)
plt.title("First Chart",color='r',fontsize=25)
plt.xlabel('Roll Number')
plt.ylabel('Marks')
plt.ylim([0,100])
plt.xlim([1,10])
plt.xticks(np.arange(1,11,1))
plt.yticks(np.arange(0,110,10))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[10,40,30,80,40]
plt.plot(x,y)
plt.title("First Chart",color='r',fontsize=25)
plt.xlabel("Roll Number")
plt.ylabel("Marks")
plt.ylim([0,100])
plt.xlim([1,10])
plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,100,10))
plt.legend(['DSP'])
plt.show()
```
