# Pandas_Pivot_and_melt_Function

```python
import pandas as pd
from pandas import DataFrame
df1=DataFrame(
              [
                  ['Aryan','Singh','Full-time','adminstration',2,20000],
                  ['Rohan','Agarawal','intern','Technical',3,5000],
                  ['Riya','Shah','Full-time','adminstration',5,10000],
                  ['Yash','Bhatia','Part_time','Technical',7,10000],
                  ['Siddhant','Khanna','Full-time','Management',6,20000],
              ],columns=['First Name','Last Name','Type','Department','YoE','Salary']
            )
df1
```

```python
table=pd.pivot_table(df1,index="Type")
table
```

```python
table=pd.pivot_table(df1,index="Type",values="Salary")
table
```

```python
p_table=pd.pivot_table(df1,index="Type",columns="Department",values="Salary")
p_table
```

```python
table=pd.pivot_table(df1,index="Type",columns="Department",values="Salary",aggfunc="sum")
table
```

# Pivot_table_exercise_titanic

```python
import pandas as pd
import numpy as np
df=pd.read_csv("titanic.csv")
df
```

```python
res=df.pivot_table(values=['Survived'],index=['Pclass','Sex'],aggfunc='sum')
res
```

```python
sur=df.pivot_table(values='Survived',index='Sex',columns='Pclass',aggfunc='count')
sur
```

```python
sur1=df.pivot_table(values='Survived',index='Sex',columns='Pclass',aggfunc='sum')
sur1
```

```python
sur3=df.pivot_table(values='Survived',index='Sex',columns='Pclass')
sur3
```

```python
rate_by_gender=df.pivot_table(values='Survived',index='Sex')
rate_by_gender
```

```python
result=df.pivot_table('Survived',index=['Sex','Age'],columns='Pclass')
result
```

```python
age=pd.cut(df['Age'],[0,10,30,60,80])
age
```

```python
sur_age_range=df.pivot_table('Survived',index=['Sex',age],columns='Pclass',aggfunc='count')
sur_age_range
```
